# Following Code Shows the No.of Cars Sold by Each Company Year Wise.

```python
result_df = df.groupby(['company', df['date'].dt.year])['car_id'].count().reset_index()
result_df.columns = ["company",'date','no_of_cars_sold']
result_df.head(10)
```

```python
result_df = result_df.sort_values(by = ["date","no_of_cars_sold"],ascending =False)
result_df[25:35]                   #displaying no.of cars sold by each company in each year in descending order by date and no_of_cars_sold
```

```python
# Get unique years present in the DataFrame
unique_years = result_df['date'].unique()

# Iterate over each year and find the company with max cars sold
for year in unique_years:
    max_cars_company = result_df[result_df['date'] == year].groupby('company')['no_of_cars_sold'].max().idxmax()
    max_cars_count = result_df[result_df['date'] == year].groupby('company')['no_of_cars_sold'].max().max()
    print(f"Year: {year}, Company with the most cars sold: {max_cars_company}, Number of cars sold: {max_cars_count}")

```
